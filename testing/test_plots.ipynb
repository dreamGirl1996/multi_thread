```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Throughput vs. number of cores (2 curves)

```python
throughput_pr = np.array([95.4576, 118.433, 164.041])
throughput_tp = np.array([7.64978, 7.45805, 7.80344])
core = np.array([1., 2., 4.])

plt.figure(figsize=(10, 6))
plt.plot(core, throughput_pr, '.--', label='thread per request')
plt.plot(core, throughput_tp, '.--', label='thread pool')
plt.legend()
plt.xlabel('number of cores')
plt.ylabel('throughput (req/s)')
plt.title('Throughput vs. number of cores')
plt.show()
```
