# Plots based on maximal request number of 1024

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Throughput vs. number of cores (2 curves)

```python
throughput_pr = np.array([71.0553, 115.642, 129.282])
throughput_tp = np.array([7.45286, 7.84361, 8.0032])
core = np.array([1., 2., 4.])

plt.figure(figsize=(10, 6))
plt.plot(core, throughput_pr, '.--', label='thread per request')
plt.plot(core, throughput_tp, '.--', label='thread pool')
plt.legend()
plt.xlabel('number of cores')
plt.ylabel('throughput (req/s)')
plt.title('Throughput vs. number of cores')
plt.show()
```

## Throughput vs. variations in delay (2 curves)

```python
throughput_pr = np.array([142.314, 26.0762])
throughput_tp = np.array([7.82676, 1.45316])
core = np.array([0,1])

plt.figure(figsize=(10, 6))
plt.plot(core, throughput_pr, '.--', label='thread per request')
plt.plot(core, throughput_tp, '.--', label='thread pool')
plt.legend()
plt.xlabel('variations in delay, 0 represents small variation and 1 represents large variation')
plt.ylabel('throughput (req/s)')
plt.title('Throughput vs. variations in delay')
plt.show()
```
